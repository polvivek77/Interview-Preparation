## 🐍 Python Basics (1–15)

```python
# 1. Reverse a string
s = "hello"
print(s[::-1])
```

```python
# 2. Check if a string is palindrome
s = "madam"
print(s == s[::-1])
```

```python
# 3. Find the largest number in a list
nums = [10, 5, 20, 8]
print(max(nums))
```

```python
# 4. Find the smallest number in a list
print(min(nums))
```

```python
# 5. Sum of elements in a list
print(sum(nums))
```

```python
# 6. Count vowels in a string
s = "hello"
count = sum(1 for ch in s if ch in "aeiouAEIOU")
print(count)
```

```python
# 7. Check even or odd
n = 10
print("Even" if n % 2 == 0 else "Odd")
```

```python
# 8. Swap two numbers
a, b = 5, 10
a, b = b, a
print(f"a = ", a)
print(f"b = ", b)
```

```python
# 9. Find factorial of a number
import math
print(math.factorial(5))
```

```python
# 10. Factorial without library
fact = 1
for i in range(1, 6):
    fact *= i
print(fact)
```

```python
# 11. Check prime number
n = 7
is_prime = n > 1 and all(n % i != 0 for i in range(2, n))
print(is_prime)
```

```python
# 12. Print Fibonacci series
a, b = 0, 1
for _ in range(5):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# 13. Count characters in string
s = "hello"
print(len(s))
```

```python
# 14. Convert string to lowercase
print("HELLO".lower())
```

```python
# 15. Remove spaces from string
s = "hello world"
print(s.replace(" ", ""))
```

## 📋 Lists & Tuples (16–30)

```python
# 16. Remove duplicates from list
nums = [5, 9, 1, 2, 2, 3]
print(list(set(nums)))
```

```python
# 17. Find common elements in two lists
a = [1, 2, 3]
b = [2, 3, 4]
print(list(set(a) & set(b)))
```

```python
# 18. Sort a list
nums1 = [6, 8, 5, 9, 1, 2, 2, 3]
nums1.sort()
print(nums1)
```

```python
# 19. Second largest number
nums = [10, 20, 30]
nums.sort()
print(nums[-2])
```

```python
# 20. Check list is empty
lst = []
print(len(lst) == 0)
```

```python
# 21. Flatten a list
lst = [[1,2],[3,4]]
flat = [x for sub in lst for x in sub]
print(flat)
```

```python
# 22. Reverse a list
nums = [5,8,6,3,2,9,7]
nums.reverse()
print(nums)
```

```python
# 23. Count occurrences in list
nums = [1,2,2,3]
print(nums.count(2))
```

```python
# 24. Convert list to tuple
t = tuple(nums)
print(t)
```

```python
# 25. Find index of element
print(nums.index(3))
```

```python
# 26. Merge two lists
a = [1, 2, 3]
b = [2, 3, 4]
c = a + b
print(c)
```

```python
# nums.remove(2)
nums.remove(2)
print(nums)
```

```python
# 28. Find length of list
print(len(nums))
```

```python
# 29. Copy a list
new_list = nums.copy()
print(new_list)
```

```python
# 30. Check if element exists
print(3 in nums)
```

## 📘 Dictionary & Set (31–40)

```python
# 31. Create a dictionary
d = {"a": 1, "b": 2}
print(d)
```

```python
# 32. Access dictionary value
print(d["a"])
```

```python
# 33. Get all keys
print(d.keys())
```

```python
# 34. Get all values
print(d.values())
```

```python
# 35. Count word frequency
s = "hello hello world"
freq = {}
for word in s.split():
    freq[word] = freq.get(word, 0) + 1
print(freq)
```

```python
# 36. Merge two dictionaries
d1 = {"a":1}
d2 = {"b":2}
d1.update(d2)
print(d1)
```

```python
# 37. Check key exists
print("a" in d)
```

```python
# 38. Remove key from dictionary
d.pop("a")
print(d)
```

```python
# 39. Create a set
s = {1, 2, 3}
print(s)
```

```python
# 40. Set union
print({1,2} | {2,3})
```

## ⚙️ Functions & Logic (41–50)

```python
# 41. Simple function
def add(a, b):
    return a + b

add(8,6)
```

```python
# 42. Lambda function
addd = lambda a, b: a + b

addd(77,2)
```

```python
# 43. Default argument
def greet(name="User"):
    print(name)
```

```python
# 44. Keyword arguments
```

```python
add(b=2, a=3)
```

```python
# 45. *args example
def total(*nums):
    return sum(nums)

nums = [2,5,6]
total(*nums)
```

```python
# 46. **kwargs example
def info(**data):
    print(data)
    
info(city="New York", job="Engineer")
```

```python
# 47. Try-except block
try:
    print(10/0)
except ZeroDivisionError:
    print("Error")
```

```python
# 48. Read input
x = int(input("Enter number: "))
print(x)
```

```python
# 49. List comprehension
squares = [x*x for x in range(5)]
print(squares)
```

```python
# 50. Check type of variable
print(type(10))
```
